```python
import numpy as np
import pandas as pd

df = pd.read_csv('tfidf_output.csv')

jumlah_baris, jumlah_kolom = df.shape

print("Jumlah baris dalam dataset:", jumlah_baris)
print("Jumlah kolom dalam dataset:", jumlah_kolom)
print('')
print(df)
U, S, Vt = np.linalg.svd(df)

U_matrix = pd.DataFrame(U, index=df.index)  
singular_values_vector = pd.DataFrame(S, columns=['Singular Values'])  
Vt_matrix = pd.DataFrame(Vt, columns=df.columns) 
```

```python
import pandas as pd
import numpy as np
from scipy.linalg import svd

# Load the TF-IDF matrix data from the uploaded file
file_path = 'tfidf_output.csv'
tfidf_matrix = pd.read_csv(file_path)

# Perform Singular Value Decomposition (SVD)
U, singular_values, Vt = svd(tfidf_matrix.values)

# Prepare results: U, singular values (Sigma), and V transpose
U_matrix = pd.DataFrame(U, index=tfidf_matrix.index)  # U matrix with original index
singular_values_vector = pd.DataFrame(singular_values, columns=['Singular Values'])  # Sigma in diagonal vector form
Vt_matrix = pd.DataFrame(Vt, columns=tfidf_matrix.columns)  # V transpose with original columns

U_matrix
```
